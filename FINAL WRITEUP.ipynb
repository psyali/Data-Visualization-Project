# 1) Why you took the approach you did?


We started with component 3 which involved generating the following components:

• A map of the United States, where the states are coloured either by the total sightings in each state over the selected time period, or the total time in sightings.

• A plot displaying the total number of sightings in whichever state is highlighted, aggregated so that it is a function of the year.

• A plot displaying the total duration of sightings in whichever state is highlighted, aggregated so that it is a function of the year.

• Tooltips should appear when hovering over the state with useful information about the state. 

Then we proceed to component 1 which involved working on the following components:

• The approach that I took for component 1 was to read in the location text file in the locations table. It then drops columns a and b which I have considered the last two redundant columns.

• Then we have set station as the index as it will be used for plotting the location of the different stations on the visualization.

• By setting the tohoku location and using the haversine formula, I have calculated the distance of each location from the tohoku location.

• This distance values are then sorted.

• The data_tohoku_norm_transpose data is read in the array_vals data frame and a time range of 4 hours is set with each observation differing by 1 sec.

• Normalization by replacing the NaN values by the mean is done

• Functions are written then to plot the spectrogram, slider, scatterplot.

• The interactions between all these functional graphs are done by importing libraries like boken,bqplot, IPython.display and many others.
